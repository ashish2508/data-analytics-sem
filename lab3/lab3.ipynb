```python
#Q1 Print the multiplication table of a number.
n = int(input("Enter a number: "))

for i in range(1, 11):
    print(n, "x", i, "=", n * i)

```

```python
#Q2: Print all even numbers between 1 and 100
for i in range(2, 101, 2):
    print(i)
```

```python
#Q3 Print sum of all numbers from 1 to N.
n = int(input("Enter N: "))
s = 0

for i in range(1, n + 1):
    s += i

print("Sum =", s)
```

```python
#Q4: Calculate factorial of a number.
n = int(input("Enter a number: "))
fact = 1

for i in range(1, n + 1):
    fact *= i

print("Factorial =", fact)
```

```python
#Q5:Count digits in a number.
n = int(input("Enter a number: "))
count = 0

while n != 0:
    count += 1
    n //= 10

print("Number of digits =", count)
```

```python
l = []
n = int(input("Enter the number of elements in list: "))
for i in range(n):
    l.append(int(input(f"Enter the {i}th element: ")))
    
print("\nOriginal list:", l)

# Insert
x = int(input("\nEnter index to insert: "))
l.insert(x, int(input("Enter element: ")))
print("After insert:", l)

# Sort
l.sort()
print("\nAfter sort:", l)

# Pop
l.pop()
print("After pop:", l)

# Reverse
l.reverse()
print("\nAfter reverse:", l)

# Extend
l2 = [int(input(f"Enter element {i} to extend: ")) for i in range(int(input("\nElements to extend: ")))]
l.extend(l2)
print("After extend:", l)

# Reverse sort
l.sort(reverse=True)
print("\nReverse sorted:", l)

# Index
print("Index of", (elem := int(input("\nFind element: "))), "is", l.index(elem))

# Remove
l.remove(int(input("\nRemove element: ")))
print("After remove:", l)

# Delete
del l[int(input("\nDelete at index: "))]
print("After delete:", l)

# Clear
if input("\nClear list? (yes/no): ").lower() == 'yes':
    l.clear()
print("Final list:", l)
```
