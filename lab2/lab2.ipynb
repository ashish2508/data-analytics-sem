```python
def power(base, exp):
    """
    Compute base raised to the power exp.

    Parameters:
        base (int or float): The number to be raised.
        exp (int): The exponent (must be an integer).

    Returns:
        float: The result of base ** exp.

    Raises:
        TypeError: If base or exp are not numeric types.
        ValueError: If exponent is not an integer.
    """

    if not isinstance(base, (int, float)):
        raise TypeError("base must be int or float")
    if not isinstance(exp, int):
        raise ValueError("exp must be an integer")

    if exp == 0:
        return 1

    result = 1
    if exp > 0:
        for _ in range(exp):
            result *= base
    else:
        for _ in range(-exp):
            result *= base
        result = 1 / result

    return result
```
