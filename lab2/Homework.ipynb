```python
import statistics as st
import numpy as np


data=[600,470,170,430,300]
mean_value=st.mean(data)
print('Mean value:',mean_value)

median_value=st.median(data)
print('Median value',median_value)

variance_value=st.variance(data)
print('Variance:',variance_value)

sd=st.pstdev(data)
print('Standard Devitiation:',sd)

maximum=max(data)
minimum=min(data)
range=maximum-minimum 
print('Range:',range)

q1,q2,q3 = np.quantile(data, [0.25, 0.50, 0.75]) 

print("Q1:", q1)
print("Q2 (Median):", q2)
print("Q3:", q3)
```

```python
import statistics as st
import numpy as np
import matplotlib.pyplot as plt

data=[15,29,8,42,35,21,18,42,26]
mean_value=st.mean(data)
print('Mean value:',mean_value)

median_value=st.median(data)
print('Median value',median_value)

variance_value=st.variance(data)
print('Variance:',variance_value)

sd=st.pstdev(data)
print('Standard Devitiation:',sd)

maximum=max(data)
minimum=min(data)
range=maximum-minimum 
print('Range:',range)

q1 = np.quantile(data, 0.25) 
q2 = np.quantile(data, 0.50)
q3 = np.quantile(data, 0.75)

print("Q1:", q1)
print("Q2 (Median):", q2)
print("Q3:", q3)

plt.hist(data, bins=5, edgecolor='black')


plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(median_value, color='green', linestyle='dashed', linewidth=2, label='Median')

plt.title("Histogram of Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()

plt.show()
```
