```python
#q1
lst = [4, 2, 9, 1, 7]

max_val = lst[0]
min_val = lst[0]

for x in lst:
    if x > max_val:
        max_val = x
    if x < min_val:
        min_val = x

print("Max:", max_val)
print("Min:", min_val)

```

```python
#q2
lst = [1, 2, 3, 4, 5, 6]
even = 0
odd = 0

for x in lst:
    if x % 2 == 0:
        even += 1
    else:
        odd += 1

print("Even:", even)
print("Odd:", odd)
```

```python
#q3
lst = [1, 2, 2, 3, 4, 4, 5]
unique = []

for x in lst:
    if x not in unique:
        unique.append(x)

print(unique)
```

```python
#q4
lst = [10, 20, 4, 45, 99]

largest = second = -1

for x in lst:
    if x > largest:
        second = largest
        largest = x
    elif x > second and x != largest:
        second = x

print("Second largest:", second)
```

```python
#q5
lst = [5, 3, 8, 4, 2]
n = len(lst)

for i in range(n):
    for j in range(0, n - i - 1):
        if lst[j] > lst[j + 1]:
            lst[j], lst[j + 1] = lst[j + 1], lst[j]

print(lst)
```

```python
#q9
lst = [1, 2, 3, 4, 5]
is_sorted = True

for i in range(len(lst) - 1):
    if lst[i] > lst[i + 1]:
        is_sorted = False
        break

print(is_sorted)
```

```python
#q11
sentence = "apple banana apple mango banana apple"
words = sentence.split()
freq = {}

for w in words:
    if w in freq:
        freq[w] += 1
    else:
        freq[w] = 1

print(freq)
```
