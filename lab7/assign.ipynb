```python
#dataset
import numpy as np
import pandas as pd

np.random.seed(42)

n = 500

data = pd.DataFrame({
    "age": np.random.randint(18, 60, n),
    "income": np.random.normal(50000, 12000, n),  
    "expenses": np.random.normal(30000, 8000, n),
    "sales": np.random.exponential(20000, n) 
})

print(data.head())
```

```python
#q1
col = data["income"]

mean = col.mean()
median = col.median()
mode = col.mode()[0]
variance = col.var()
std_dev = col.std()

print("Mean:", mean,"\nMedian:", median,"\nMode:", mode,"\nVariance:", variance,"\nStandard Deviation:", std_dev)
```

```python
#q2
data["income_zscore"] = (data["income"] - data["income"].mean()) / data["income"].std()

print(data[["income", "income_zscore"]].head())
```

```python
#q3
data["income_minmax"] = (data["income"] - data["income"].min()) / \
                        (data["income"].max() - data["income"].min())

print(data[["income", "income_minmax"]].head())
```

```python
#q4
sample = np.random.normal(loc=50, scale=10, size=100)

print("Sample size:", len(sample))
print("Sample mean:", np.mean(sample))
```

```python
#q5 
A_and_B = 30
B = 50

P_A_given_B = A_and_B / B

print("P(A|B):", P_A_given_B)
```

```python
#q6
correlation = data["income"].corr(data["expenses"])

print("Correlation:", correlation)

```

```python
#q7
train_ratio = 0.8

shuffled = data.sample(frac=1, random_state=42).reset_index(drop=True)

train_size = int(len(data) * train_ratio)

train = shuffled[:train_size]
test = shuffled[train_size:]

print("Train size:", len(train))
print("Test size:", len(test))
```

```python
#q8
print("Skewness before:", data["sales"].skew())
data["sales_log"] = np.log1p(data["sales"])   # log(1+x) to avoid log(0)

print("Skewness after:", data["sales_log"].skew())
```
