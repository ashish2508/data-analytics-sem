```python
import pandas as pd
```

```python
#q1
df = pd.read_csv("data.csv")
col = "marks"

mean_val = df[col].mean()
result = df[df[col] > mean_val]
print(result)
```

```python
#q2
print(df.isnull().sum())
```

```python
#q3
for c in df.columns:
    if df[c].dtype in ["int64", "float64"]:
        df[c].fillna(df[c].mean(), inplace=True)
    else:
        df[c].fillna(df[c].mode()[0], inplace=True)
```

```python
#q4
df.rename(columns={"old_name": "new_name"}, inplace=True)
df.drop(columns=["unwanted_col1", "unwanted_col2"], inplace=True)
```

```python
#q5
# AND
filtered_and = df[(df["age"] > 25) & (df["marks"] > 60)]

# OR
filtered_or = df[(df["age"] > 25) | (df["marks"] > 60)]
```
