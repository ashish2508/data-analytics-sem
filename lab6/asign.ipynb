```python
import pandas as pd
```

```python
#q1
df = pd.read_csv("data.csv")
col = "marks"

mean_val = df[col].mean()
result = df[df[col] > mean_val]
print(result)
```

```python
#q2
print("Missing Values: \n",df.isnull().sum())
```

```python
#q3
for c in df.columns:
    if df[c].dtype in ["int64", "float64"]:
        df[c] = df[c].fillna(df[c].mean())
    else:
        df[c] = df[c].fillna(df[c].mode()[0])

```

```python
#q4
df.rename(columns={'marks': 'Marks'}, inplace=True)
df.drop(columns=['id'], inplace=False)

print(df.columns)
```

```python
#q5
_and = df[(df["age"] > 20) & (df["Marks"] > 60)]
_or = df[(df["age"] > 20) | (df["Marks"] > 60)]

print("Using AND - ", _and.shape[0])
print("Using OR - ", _or.shape[0])


```

```python
#q6
df["BMI"] = df["weight"] / (df["height"] ** 2)

print(df["BMI"].head())

```

```python
#q7
unique_values = df["gender"].unique()
label_map = {value: index for index, value in enumerate(unique_values)}

df["gender_encoded"] = df["gender"].map(label_map)

print(label_map)
print(df["gender_encoded"].head())
```

```python
#q8
grouped = df.groupby("department")["Marks"].agg(["mean", "max", "min", "count"])

print(grouped)
```

```python
#q9
df_sorted = df.sort_values(by=["age", "Marks"], ascending=[True, False])

print(df_sorted[["age", "Marks"]].head())
```

```python
#q10
df_extra = df[["id", "department"]].copy()

for join_type in ["inner", "outer", "left", "right"]:
    merged = pd.merge(df, df_extra, on="id", how=join_type)
    print(f"{join_type.capitalize()} shape:", merged.shape)
```

```python
#q11
pivot = pd.pivot_table(df,values="Marks",index="department",columns="gender",aggfunc="mean"
)

print(pivot)

```

```python
#q12
Q1 = df["Marks"].quantile(0.25)
Q3 = df["Marks"].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df["Marks"] < lower) | (df["Marks"] > upper)]

print("Number of outliers:", outliers.shape[0])

```

```python
#q13
before = df.shape[0]

df.drop_duplicates(inplace=True)

after = df.shape[0]

print("Rows removed:", before - after)

```

```python
#q14
df.to_csv("Cars_data_cleaned.csv", index=False)

print("File saved successfully")

```
